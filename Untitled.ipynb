```python
# Output verisinin indexleri
outputs = {
    0: "Excellent",
    1: "Good",
    2: "Fair",
    3: "Poor"
}
outputs_oppo = {
    "Excellent": 0,
    "Good": 1,
    "Fair": 2,
    "Poor": 3
}
```

```python
# Veri okunacak;
import pandas as pd
df = pd.read_excel('cellular_quality.xlsx').fillna(0)
df
```

```python
# train_data ve train_labels hazırlanacak;
import numpy as np
train_data   = np.zeros((len(df.values), 6))
train_labels = np.zeros((len(df.values)))

for i in range(len(df.values)):
    for j, element in enumerate(df.values[i][:6]):
        train_data[i][j] = abs(element)
    # train_labels üzerinde normalizasyon yapılacak;
    train_labels[i] = int(outputs_oppo[df.values[i][6]])
```

```python
train_data
```

```python
train_labels
```

```python
# train_data üzerinde normalizasyon yapılacak;
mean = train_data.mean(axis=0)
train_data -= mean
std = train_data.std(axis=0)
train_data /= std
```

```python
train_data.mean(axis=0)[0]
```

```python
train_data[0]
```

```python
train_data_partial_length = int(len(train_data)*80/100)
train_data_partial_length
```

```python
# Train ve Test datası ayarlanacak;
x_train_partial = train_data[:train_data_partial_length]
y_train_partial = train_labels[:train_data_partial_length]
x_val = train_data[train_data_partial_length:]
y_val = train_labels[train_data_partial_length:]

# x_val = train_data[:train_data_partial_length]
# y_val = train_labels[:train_data_partial_length]
# x_train_partial = train_data[train_data_partial_length:]
# y_train_partial = train_labels[train_data_partial_length:]
```

```python
# Model Defination
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential()
model.add(layers.Dense(64, activation = "relu", input_shape = (train_data.shape[1],)))
model.add(layers.Dense(64, activation = "relu"))
model.add(layers.Dense(1))
```

```python
# Compiling the model
from keras import optimizers
# optimizer = optimizers.RMSprop(lr=0.001)
model.compile(optimizer = "rmsprop",
              loss = "mse",
              metrics = ["acc"])
```

```python
# Train your model
history = model.fit(x_train_partial,
                    y_train_partial,
                    epochs = 100,
                    batch_size = 512, 
                    validation_data = (x_val, y_val))
```

```python
history_dict = history.history
history_dict.keys()
```

```python
# Plotting the training and validation loss
import matplotlib.pyplot as plt
history_dict = history.history
loss_values = history_dict["loss"]
val_loss_values = history_dict["val_loss"]
epochs = range(1, len(loss_values) + 1)
plt.plot(epochs, loss_values, "bo", label = "training loss")
plt.plot(epochs, val_loss_values, "b", label = "validation loss")
plt.title("Training and validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
# Plotting the training and validation accuary
plt.clf()

acc = history_dict["acc"]
val_acc = history_dict["val_acc"]
plt.plot(epochs, acc, "mo", label = "Training Accuracy")
plt.plot(epochs, val_acc, "c", label="Validation Accuracy")
plt.title("Training & Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()

plt.show()
```

```python
model.evaluate(x_val, y_val)
```
